Sensor which are placed in road junction collect the data of no of vehicles at different junction and gives the data to the transport manager.

Now our task is to predict the total no of vehicles based on sensor data.

This article explain how to deal with sensor data given with a timestamp
and predict the count of vehicles at a particular time .

Approach:


*   import necessary modules

*   laod the dataset

*   Analyse the data

*   convert DateTime attribute to week , days , hours , month etc

*   Build the model


*   Train the model


*  Test the data


*  Predict the results



```python
# importing the pandas module to loading the dataframe.
import pandas as pd
df=pd.read_csv('/content/vehicles.csv')
df
```

This dataset contain 2 attributes.They are Datetime and vehicles.

Where vehicles is class Label.

The class label is of numeric type. So the regression technique is well suited for this problem. Regression is used to map the data into a predefined function it is a supervised Machine learning algorithm that is used to predict the value based on historical data. We can perform regression on our data if the data is numeric. Here class label Ie Vehicles attribute is the class label which is numeric so regression should be done.

```python
# Defining the function for getting month , day ,hours
# from the datetime and loading it to differrent columns

#get date
def get_dom(dt):
  return dt.day

#get weekday
def get_weekday(dt):
  return dt.weekday()

#get hour
def get_hour(dt):
  return dt.hour

#get year
def get_year(dt):
  return dt.year

#get month
def get_month(dt):
  return dt.month

#get year day
def get_dayofyear(dt):
  return dt.dayofyear

#get year week
def get_weekofyear(dt):
  return dt.weekofyear


df['DateTime']=df['DateTime'].map(pd.to_datetime)
df['date']=df['DateTime'].map(get_dom)
df['weekday']=df['DateTime'].map(get_weekday)
df['hour']=df['DateTime'].map(get_hour)
df['month']=df['DateTime'].map(get_month)
df['year']=df['DateTime'].map(get_year)
df['year']=df['DateTime'].map(get_year)
df['dayofyear']=df['DateTime'].map(get_dayofyear)
df['weekofyear']=df['DateTime'].map(get_weekofyear)

df.head()
```

Feature Selection and Model Developmennt

```python
# there is no use of Datetime module so we will remove it
# and remove class label
feature=df.drop(['DateTime','Vehicles'],axis=1)

# store the class label in target columns
target=df['Vehicles']

print(feature.head())
print()
print(target.head())

```

As we have modified the Dateframe from DateTime format ,we will not using Train test split module

```python
# splliting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(feature,target,random_state=42,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')

# initailising the regressor
RFR=RandomForestRegressor(n_estimators=100,max_depth=20)

RFR.fit(x_train,y_train)

```

The model takes the list of feature provided as input set and tries to predict the target value.

```python
from  sklearn.metrics import r2_score
y_pred=RFR.predict(x_train)
print('Accuracy ',r2_score(y_train,y_pred))
```

```python
#making prediction
RFR.predict([[1,6,0,11,2015,305,44]])
```
